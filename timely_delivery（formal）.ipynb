```python
import pandas as pd
file_path = '/content/01.timeline.xlsx'
df = pd.read_excel(file_path, sheet_name='sheet1')
print(df.head())
display(df.head())
```

```python
print(df.info())
```

```python
import pandas as pd
df['订单创建日期'] = pd.to_datetime(df['订单创建日期'], errors='coerce')
df = df.dropna(subset=['订单创建日期'])
daily_order_count = df.groupby('订单创建日期').size().reset_index(name='order_count')
daily_order_count = daily_order_count.sort_values(by='订单创建日期')
print(daily_order_count)
```

```python
df['送达时间'] = df['送达时间'].str.split(' ').str[0]
df['送达日期'] = pd.to_datetime(df['送达时间']).dt.date
print(df[['送达时间', '送达日期']].head())
```

```python
delivery_counts = df.groupby('送达日期').size().reset_index(name='送达数量')

# Print the result
print(delivery_counts)
```

```python
na_count = df['送达时间'].isna().sum()
print(f"'送达时间' 列中的 N/A 数量为: {na_count}")
```

```python
import pandas as pd
df['送达时间'] = pd.to_datetime(df['送达时间']).dt.tz_localize(None)
df['订单创建时间'] = pd.to_datetime(df['订单创建时间']).dt.tz_localize(None)
df['Difference'] = df['送达时间'] - df['订单创建时间']
display(df[['订单创建时间','送达时间', 'Difference']].head())
```

```python
df_filtered = df[df['Difference'].dt.days < 1]
daily_order_count = daily_order_count.sort_values(by='订单创建日期')
print(daily_order_count)
```

```python
df_filtered = df[df['Difference'].dt.days < 1]
df_filtered['订单创建日期'] = df_filtered['订单创建时间'].dt.date
daily_order_count = df_filtered.groupby('订单创建日期').size().reset_index(name='时效小于1天的订单数量')
daily_order_count = daily_order_count.sort_values(by='订单创建日期')
print(daily_order_count)
```

```python
df_filtered = df[(df['Difference'].dt.days >= 1) & (df['Difference'].dt.days < 2)]
df_filtered['订单创建日期'] = df_filtered['订单创建时间'].dt.date
daily_order_count = df_filtered.groupby('订单创建日期').size().reset_index(name='时效大于等于1天且小于2天的订单数量')
daily_order_count = daily_order_count.sort_values(by='订单创建日期')
print(daily_order_count)
```

```python
df_filtered = df[df['Difference'].dt.days > 2]
df_filtered['订单创建日期'] = df_filtered['订单创建时间'].dt.date
daily_order_count = df_filtered.groupby('订单创建日期').size().reset_index(name='时效大于2天的订单数量')
daily_order_count = daily_order_count.sort_values(by='订单创建日期')
print(daily_order_count)
```

```python
df_filtered = df[(df['Difference'].dt.days >= 0) & (df['Difference'].dt.days <= 5)]
df_filtered['订单创建日期'] = df_filtered['订单创建时间'].dt.date
daily_order_count = df_filtered.groupby('订单创建日期').size().reset_index(name='时效0-5天的订单数量')
daily_order_count = daily_order_count.sort_values(by='订单创建日期')
print(daily_order_count)
```

```python
df_filtered = df[(df['Difference'].dt.days >= 0) & (df['Difference'].dt.days <= 7)]
df_filtered['订单创建日期'] = df_filtered['订单创建时间'].dt.date
daily_order_count = df_filtered.groupby('订单创建日期').size().reset_index(name='时效0-7天的订单数量')
daily_order_count = daily_order_count.sort_values(by='订单创建日期')
print(daily_order_count)
```

```python
import pandas as pd
df['Difference'] = df['送达时间'] - df['订单创建时间']
df_filtered = df[(df['Difference'].dt.days <= 5)]
operator_time_efficiency = df_filtered.groupby('物流方式').size().reset_index(name='每个运营商时效在5天的订单数量')
operator_time_efficiency = operator_time_efficiency.sort_values(by='每个运营商时效在5天的订单数量', ascending=False)
display(operator_time_efficiency)
```
